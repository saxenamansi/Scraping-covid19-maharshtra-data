## Importing necessary libraries

```python
import pandas as pd
import requests
```

## Reading the html content from the webpage and scraping all tables 

```python
url = 'https://en.wikipedia.org/wiki/COVID-19_pandemic_in_Maharashtra'
```

```python
dfs = pd.read_html(url) #will read all tables
```

```python
print("Total number of tables scraped: ", len(dfs)) #total number of tables available in the given url
```

## Displaying the three tables required and storing them in excel files

### Table 1

![title](table1.png)

```python
table1 = dfs[0]
table1 = table1[2:-2]
table1.reset_index(drop = True, inplace = True)
table1.to_excel('scraping_table1.xlsx')
table1
```

### Table 2

![title](table2.png)

```python
table2 = dfs[1]
table2.columns = table2.iloc[0]
table2 = table2[1:-1]
table2.reset_index(drop = True, inplace = True)
table2.to_excel('scraping_table2.xlsx')
table2
```

### Table 3

![title](table3.png)

```python
table3 = dfs[4]
table3 = table3[:-1]
table3.to_excel('scraping_table3.xlsx')
table3
```
